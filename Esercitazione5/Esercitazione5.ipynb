## Step con distribuzione di probabilit√† uniforme

psi100

$\delta=0.9$

$\langle \alpha \rangle=0.523535$

```python
M=1000000  #numero step totali
N=100  #numero blocchi
xb=np.arange(N)+1  #ascisse per plot raggio medio

ru1, erroru1=np.loadtxt("unif1rad.dat", usecols=(0, 1),  unpack="true")  #array per plot valor medio di r
xu1, yu1, zu1=np.loadtxt("unif1pos.dat", usecols=(0, 1, 2),  unpack="true")  #array per plot posizioni in 3d

u1fig ,(u1plt1, u1plt2) = plt.subplots(1, 2, figsize=(15, 5))

u1plt1.errorbar(xb, ru1, yerr=erroru1)
u1plt1.axhline(y=3/2, color="mediumvioletred", label="Valore atteso", alpha=0.7)
u1plt1.set_xlabel("Blocchi")
u1plt1.set_ylabel(r"$\left\langle \frac{r}{a_0} \right\rangle_{\Psi_{1,0,0}}$")
u1plt1.set_title("Media a blocchi del raggio in unit√† di raggi di Bohr")
u1plt1.grid(True)

u1plt2.axis('off')
u1plt2=u1fig.add_subplot(122, projection='3d')
u1plt2.scatter(xu1, yu1, zu1, marker='.', alpha=0.2)
u1plt2.set_xlabel('x')
u1plt2.set_ylabel('y')
u1plt2.set_zlabel('z')
u1plt2.set_title(r"Sampling di $\vert \Psi_{1,0,0} \vert ^2$ ottenuto")
u1plt2.view_init(10, 30)

u1fig.suptitle(r"Analisi relativa a $\Psi_{1,0,0}$ ottenuta mediante passi con distribuzione di probabilit√† uniforme")

plt.show()
```
